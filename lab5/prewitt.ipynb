```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

image = cv2.imread('lappy.jpg', cv2.IMREAD_GRAYSCALE)

prewitt_x = np.array([[-1, 0, 1], [-1, 0, 1], [-1, 0, 1]])
prewitt_y = np.array([[-1, -1, -1], [0, 0, 0], [1, 1, 1]])

image_x = cv2.filter2D(image, -1, prewitt_x)
image_y = cv2.filter2D(image, -1, prewitt_y)

image_prewitt = cv2.addWeighted(image_x, 0.5, image_y, 0.5, 0)

cv2.imwrite('image_prewitt.jpg', image_prewitt)

plt.imshow(image_prewitt)
```
