```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

### Write a program in python to implement gaussian blur (smoothing filter).

```python
img = cv2.imread("image.jpg")
img = img[:,:,::-1]
plt.imshow(img)
```

```python
gaussian_blur = cv2.GaussianBlur(img, (19,19), 0.9)
plt.imshow(gaussian_blur)
```

```python
k = 9
mean_filter = cv2.blur(img, (k,k))
plt.imshow(mean_filter)
```

```python
median_blur = cv2.medianBlur(img, 5)
plt.imshow(median_blur)
```

```python
adaptive_blur = cv2.adaptiveMedianBlur(img, 5)
```
