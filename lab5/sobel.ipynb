```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

image = cv2.imread('lappy.jpg', cv2.IMREAD_GRAYSCALE)

sobel_x = np.array([[-1, 0, 1], [-2, 0, 2], [-1, 0, 1]])
sobel_y = np.array([[-1, -2, -1], [0, 0, 0], [1, 2, 1]])

image_x = cv2.filter2D(image, -1, sobel_x)
image_y = cv2.filter2D(image, -1, sobel_y)

image_sobel = cv2.addWeighted(image_x, 0.5, image_y, 0.5, 0)

cv2.imwrite('image_sobel.jpg', image_sobel)

plt.imshow(image_sobel)

```
